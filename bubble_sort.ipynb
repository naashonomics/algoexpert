# Bubble Sort

Write a function to implement bubble sort 

```python
#TC: O(n^2) & SC: O(1)
def bubble_sort(array):
    for i in range(len(array)-1):
        for j in range(len(array)-1-i):
            if array[j]>array[j+1]:
                swap(array,j,j+1)
    return array

def swap(array,j,k):
    array[j],array[k]=array[k],array[j]
```

```python
array=[3,-1,4,88,77]
print(bubble_sort(array))
```

```python
array=[-1,-1,4,88,77]
print(bubble_sort(array))
```

```python
#best case is O(N) if you are given a sorted array
def bubble_sort2(array):
    issorted=False
    counter=0
    while not issorted:
        issorted=True
        for i in range(len(array)-1-counter):
            if array[i] > array[i+1]:
                issorted=False
                swap(array,i,i+1)
        counter+=1
    return array
def swap(array,i,j):
    array[i],array[j]=array[j],array[i]
    
```

```python
array=[3,-1,4,88,77]
print(bubble_sort2(array))
```
